```python
# #1. "The Toll Booth Trouble"
# At a national highway, vehicles pass through a toll booth. Each vehicle type (bike, car, truck, bus) has a different toll rate. Some vehicles have a monthly pass, so their rate is ₹0.
# Your Task:
# Design a system to compute the toll for each vehicle as it passes, using a class for each vehicle type. At the end of the day, show the total collection and number of vehicles by type.
# •	Concepts: Classes, inheritance, conditionals, dictionaries.

class Vehicle:
    def __init__(self, has_pass):
        self.has_pass = has_pass

    def get_toll(self):
        return 0 if self.has_pass else self.toll_rate

class Bike(Vehicle):
    toll_rate = 10

class Car(Vehicle):
    toll_rate = 50

class Truck(Vehicle):
    toll_rate = 100

class Bus(Vehicle):
    toll_rate = 80

class TollBooth:
    def __init__(self):
        self.total_collection = 0
        self.vehicle_count = {"Bike": 0, "Car": 0, "Truck": 0, "Bus": 0}

    def process_vehicle(self, vehicle):                # Process a vehicle passing through the toll booth
        toll = vehicle.get_toll()                    # Get the toll rate for the vehicle
        self.total_collection += toll                        # Add the toll to the total collection
        self.vehicle_count[vehicle.__class__.__name__] += 1      # Increment the count for the vehicle type

    def end_of_day_report(self):                                    
        print("Total Collection: ₹", self.total_collection)           
        for vehicle_type, count in self.vehicle_count.items():        # Iterate through the vehicle count dictionary
            print(f"{vehicle_type}: {count}")

booth = TollBooth()                   
vehicles = [Bike(False), Car(True), Truck(False), Bus(False), Car(False), Bike(True), Truck(True)]            # Create a list of vehicles passing through the toll booth

for v in vehicles:
    booth.process_vehicle(v)

booth.end_of_day_report()
```

```python
# 2. "The Library Whisperer"
# You are managing a small public library. Each book can be borrowed for up to 15 days. After that, a fine of ₹2/day is charged. Students and adults have different borrowing limits.
# Your Task:
# Design classes for User, Book, and Library. Calculate fines for overdue books, and generate a borrowing report for each user.
# •	Concepts: OOP (composition), date math, fine calculation.





from datetime import datetime, timedelta

class Book:
    def __init__(self, title):
        self.title = title
        self.borrowed_on = None

    def borrow(self):
        self.borrowed_on = datetime.now()

    def fine(self):
        if not self.borrowed_on:
            return 0
        days = (datetime.now() - self.borrowed_on).days
        return max(0, (days - 15) * 2)

class User:
    def __init__(self, name, user_type):
        self.name = name
        self.type = user_type  # 'student' or 'adult'
        self.books = []

    def borrow_book(self, book):
        limit = 3 if self.type == 'student' else 5
        if len(self.books) < limit:
            book.borrow()
            self.books.append(book)

    def report(self):
        print(f"\n{self.name}'s Report:")
        for b in self.books:
            print(f"{b.title} - Fine: ₹{b.fine()}")
        total = sum(b.fine() for b in self.books)
        print(f"Total Fine: ₹{total}")

# Demo
b1 = Book("Python")
b2 = Book("ML Guide")

u = User("Alice", "student")
u.borrow_book(b1)
b1.borrowed_on -= timedelta(days=20)  # simulate overdue

u.borrow_book(b2)
u.report()
```

```python
#  3. "The Chocolate Factory Audit"
# Willy Wonka needs your help. His chocolate machines produce candies every hour. Each type of machine produces at a different rate and has a maintenance cost.
# Your Task:
# Simulate production for a day. Use classes to represent machines. Calculate total output and operating cost. Suggest which machine is most efficient (candies per ₹).
# •	Concepts: Classes, math formulas, loops, performance calculation.




class machine():
    def __init__(self,name,rate,cost):
        self.name=name
        self.rate=rate
        self.cost=cost
    def produce(self):
        return self.rate*24 #per hr
    def m_c(self):
        return self.produce()/self.cost # Maintenace Cost Per Day
    
M1=machine("Machine A",100,10)
M2=machine("Machine B ",150,1500)
print(f"{M1.name} produces {M1.produce() } and the maintenance cost is {M1.m_c()}")
print(f"{M2.name} produces {M2.produce() } and the maintenance cost is {M2.m_c()}")
```
